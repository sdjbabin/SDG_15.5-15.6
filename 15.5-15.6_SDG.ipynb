## First let us describe what the indicator and the corresponding Value is:

- The Red List Index for a particular country or region is an index of the aggregate extinction risk for species within the country or region relative to its potential contribution to global species extinction risk (within the taxonomic groups included). It is measured on a scale of 0 to 1, where 1 is the maximum contribution that the country or region can make to global species survival, equating to all species being classified as Least Concern on the IUCN Red List, and 0 is the minimum contribution that the country or region can make to global species survival, equating to all species in the country or region having gone extinct.

<h3 align="center" style="color: brown;">Proportion of countries with performance above average</h3>

```python
time_periods = []
percentages = []

# Calculating the total number of unique countries
total_countries = df1['GeoAreaName'].nunique()

for time_period in df1['TimePeriod'].unique():
    filtered_df = df1[df1['TimePeriod'] == time_period]
    
    average_rli = filtered_df['Value'].mean()
    
    above_average_df = filtered_df[filtered_df['Value'] > average_rli]
    
    countries_above_average = above_average_df['GeoAreaName'].nunique()
    
    # Calculating the percentage for the current time period
    percentage_above_average = (countries_above_average / total_countries) * 100
    
    # Appending the time period and percentage to the lists
    time_periods.append(time_period)
    percentages.append(percentage_above_average)

# Let us create a line plot for the percentages over time
plt.figure(figsize=(12, 6))
plt.plot(time_periods, percentages,linewidth=4, linestyle='-', color='g')
plt.xlabel('Time Period',color="blue")
plt.ylabel('Percentage of Countries with RLI Above Average',color="blue")
plt.title('Percentage of Countries with RLI Above Average Over Time',color="blue")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
